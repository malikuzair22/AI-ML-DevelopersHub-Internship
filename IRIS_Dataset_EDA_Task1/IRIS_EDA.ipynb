```python
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Plot Settings
sns.set_style('darkgrid')
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# 1. Load Dataset
iris_df = pd.read_csv('/content/sample_data/IRIS.csv')
```

```python
# 2. Basic Inspection

print("Shape:", iris_df.shape)
print("\nColumn Names:", iris_df.columns.tolist())
print("\nFirst 5 Rows:")
print(iris_df.head())

print("\nDataset Info:")
iris_df.info()

print("\nDescriptive Statistics:")
print(iris_df.describe())
```

```python
# 3. Scatter Plot — Feature Relationships

plt.figure()
sns.scatterplot(data=iris_df, x='sepal_length', y='petal_length', hue='species')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.tight_layout()
plt.show()
```

```python
# 4. Histograms — Value Distributions

numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

iris_df[numeric_cols].hist(bins=20, color='steelblue', edgecolor='black', layout=(2, 2))
plt.suptitle('Feature Value Distributions', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# 5. Box Plots — Outlier Detection

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
for ax, col in zip(axes.flatten(), numeric_cols):
    sns.boxplot(data=iris_df, x='species', y=col, ax=ax, hue='species', palette='Set2', legend=False)
    ax.set_title(f'{col} by Species')
plt.suptitle('Box Plots — Outlier Detection', y=1.02)
plt.tight_layout()
plt.show()
```

```python
# 6. Pair Plot — Overall Feature Overview

sns.pairplot(iris_df, hue='species', palette='Set1')
plt.suptitle('Pair Plot of All Features', y=1.02)
plt.show()
```
